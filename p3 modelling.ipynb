# SIR MODELS OF EPIDEMICS

```python
from scipy.integrate import odeint
import numpy as np
#import matplotlib
#matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import pandas as pd



def SIR(Y,t, beta, r):
  S,I,R = Y
  dSdt= (-beta)*S*I
  dIdt= beta*S*I-r*I
  dRdt= r*I 
  return ([dSdt, dIdt, dRdt])

beta = 1e-3
r = 1e-1
S0 = 499
I0 = 1
R0 = 0
dt = np.linspace(0,100, 1000)
N = 500
Y = odeint(SIR, [S0,I0,R0], dt, args=(beta, r))
S= Y[:,0]
I= Y[:,1]
R= Y[:,2]
#N = S + I + R
plt.plot(dt, S, label='Susceptible')
plt.plot(dt, I, label='Infected')
plt.plot(dt, R, label= 'Recovered')
plt.title('SIR model')
plt.xlabel('Time')
plt.ylabel('Number of individuals')
plt.legend()
plt.show()
```

```python
r_vals =[] 
r = 1e-1
beta = 1e-3
beta_vals = []
for n in range(10):
    r = 1e-1 + (1e-1)**n
    beta = 1e-3 + (1e-3)**n
    r_vals.append(r)
    beta_vals.append(beta)
    

rs = []
bs = []
Ro_vals = []
logical_Ro = []
logical_I = []
results_I = []
for r in r_vals:
    for beta in beta_vals:
        Ro = (beta*N)/r
        Ro_vals.append(Ro)
        rs.append(r)
        bs.append(bs)
        if Ro < 1:
            logical_Ro.append('false')
        else:
            logical_Ro.append('true')
        
        Y = odeint(SIR, [S0,I0,R0], dt, args=(beta, r))
        S= Y[:,0]
        I= Y[:,1]
        R= Y[:,2]
        if max(I) > I0:
            logical_I.append('true')
        else:
            logical_I.append('false')

```
